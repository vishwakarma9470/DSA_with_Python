### Syntax, Semantics and indentation in Python

- syntax refers to the set of rules that defines the combination of symbols that are considered to be correctly structured programs in a language. In simpler terms, syntax is about the correct arrangement of word and symbols in a code.


- semantics refers to the meaning or the imterpretation of the symbols, characters and commands in a language. it is about what the code is supposed to do when it runs.


- Indentation in python is used to define the struture and hierarchy of the code. Unlike many other programming languages that use braces {} to delimit blocks of code, Python uses indentation to determine the grouping of statements. This means that all the statements within a block must be indented at the same level.

## Operators in Python


```python
## Arithmetics operators
a = 20
b = 4
print(a+b) # addition
print(a-b) # subtraction
print(a*b) # multiplication
print(a/b) # division
print(a//b) # floor division
print(a%b) # modulus
print(a**b) # exponentiation
```

```python
## comparison operators
a = 43
b = 23
print(a>b) # greater
print(a<b)  # lessthan
print(a>=b) # greater equal to
print(a<=b) # lessthan equal to
print(a==b) # equal to operation
print(a!=b) # not equal tp operation
```

```python
## logical operator
a = True
b = False
print(a and b)  # AND operation
print(a or b) # or operations
print(not a) # NOt operation
```
